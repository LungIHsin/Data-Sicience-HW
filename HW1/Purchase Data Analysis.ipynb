```python
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#Player Count
df1=purchase_data.groupby('SN')
for i in range(len(df1)):
    a=i+1
total_player=pd.DataFrame(a,columns=['Total Players'],index=[0])
display(total_player)
```

```python
#Purchasing Analysis (Total)
df2=purchase_data.groupby('Item ID')
for i in range(len(df1)):
    a=i+1
df3=purchase_data['Price']
Revenue=round(df3.mean()*df3.count(),2)
item=pd.DataFrame({'Number of Unique Items':[a],'Average Price':['$'+str(round(df3.mean(),2))],
                  'Number of Purchases':[df3.count()],'Total Revenue':['$'+str(Revenue)]})      
display(item)
```

```python
#Gender Demographics
df4=purchase_data.groupby(['SN','Gender']).count()
df4['Age']=1
df5=df4.groupby('Gender').size()
Gender=pd.DataFrame({'Percentage of Players':[round(df5[1]/a*100,2),round(df5[0]/a*100,2),round(df5[2]/a*100,2)],
                     'Total Count':[df5[1],df5[0],df5[2]]},
                    index=['Male','Female','Other / Non-Disclosed'])
display(Gender)
```

```python
#Purchasing Analysis (Gender)
df6=purchase_data.groupby('Gender')
df7=df6['Price'].agg('sum')
Purchasing=pd.DataFrame({'Purchase Count':[df6.size()[0],df6.size()[1],df6.size()[2]],
                        'Average Purchase Price':[df7[0]/df6.size()[0],df7[1]/df6.size()[1],df7[2]/df6.size()[2]],
                        'Total Purchase Value':[df7[0],df7[1],df7[2]],
                        'Avg Purchase Total per Person':['$'+str(round(df7[0]/df5[0],2)),'$'+str(round(df7[1]/df5[1],2)),'$'+str(round(df7[2]/df5[2],2))]},
                       index=['Male','Female','Other / Non-Disclosed'])
display(Purchasing)
```

```python
#Age Demographics
list1,list2=[],[]
bin=[0,9,14,19,24,29,34,39,50]
group_names=['','10-14','15-19','20-24','25-29','30-34','35-39','40+']
member=purchase_data.groupby('SN').first()
age=pd.cut(member['Age'],bin,labels=group_names)
df9=pd.DataFrame(age)
df9=df9.groupby('Age').size()
print(df9)
for i in range(len(df9)):
    percent=round(df9[i]/576*100,2)
    list1.append(percent)
    list2.append(df9[i])
Age_Demographics=pd.DataFrame({'Percentage of Players':list1,
                                'Total Count':list2},index=group_names)
display(Age_Demographics)
```

```python
#Purchasing Analysis (Age)
list3,list4,list5=[],[],[]
df8=pd.cut(purchase_data['Age'],bin,labels=group_names)
df8=pd.DataFrame(df8)
df11=df8.groupby('Age').size()
df8.rename(columns={ df8.columns[0]: "new age" }, inplace=True)
new=purchase_data.merge(df8, how='outer', left_index=True, right_index=True)
df10=new.groupby('new age')['Price'].agg('sum')
for i in range(len(df10)):
    avg_price='$'+str(round(df10[i]/df11[i],2))
    avg_person='$'+str(round(df10[i]/list2[i],2))
    price='$'+str(round(df10[i],2))
    list3.append(avg_price)
    list4.append(avg_person)
    list5.append(price)
Purchasing_Analysis=pd.DataFrame({'Purchase Count':df11,'Average Purchase Price':list3,
                                 'Total Purchase Value':list5,'Average Purchase Total per Person':list4},
                                 index=group_names)
display(Purchasing_Analysis)
```

```python
#Top Spenders
list6=[]
purchase=purchase_data.groupby('SN')
for i in purchase:
    list6.append(i[0])
sum1=purchase['Price'].agg('sum')
count1=purchase['Price'].size()
df12=pd.DataFrame({'Purchase Count':count1,
                  'Average Purchase Price':round(sum1/count1,2),
                  'Total Purchase Value':sum1},
                  index=list6)
df12.rename_axis('SN',inplace=True)
df12=df12.sort_values(by='Total Purchase Value',ascending = False).head(5)
display(df12)
```
