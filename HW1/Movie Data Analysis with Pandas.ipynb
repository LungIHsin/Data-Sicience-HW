```python
#Homework 1
#第一題
import pandas as pd
file=pd.read_csv('IMDB-Movie-Data.csv')
df=pd.DataFrame(file)
df=df.fillna(0) 
df['Genre']=df['Genre'].str.split('|')
df['Actors']=df['Actors'].str.split('|')
#display(df)

#第一小題
def highest_rating():
    df7=df[df.Year==2016]
    df7=df7.sort_values('Rating', ascending=False)
    return df7.head(3)

#第二小題
def  highest_average_revenue():
    actor_dict={}
    for m in df.index:
        for actor in df.iloc[m]['Actors']:
            if actor in actor_dict.keys():
                    actor_dict[actor][0]+=df.iloc[m]['Revenue (Millions)']
                    actor_dict[actor][1]+=1
            else:
                actor_dict[actor]=[df.iloc[m]['Revenue (Millions)'],1]
    for actor in actor_dict:
            average=actor_dict[actor][0]/actor_dict[actor][1]
            actor_dict[actor]=average

    df1=pd.DataFrame(list(actor_dict.items()))
    return df1.loc[df1[1].idxmax()]

#第三小題
def emma_average_rating():
    emma_dict={}
    for m in df.index:
        for actor in df.iloc[m]['Actors']:
            if actor== 'Emma Watson':
                if 'Emma Watson' in emma_dict.keys():
                    emma_dict[actor][0]+=df.iloc[m]['Rating']
                    emma_dict[actor][1]+=1
                else:
                    emma_dict[actor]=[df.iloc[m]['Rating'],1]

    avg=emma_dict['Emma Watson'][0]/emma_dict['Emma Watson'][1]
    emma_avg={'Emma Watson':avg}
    df2=pd.DataFrame(emma_avg,index=['average rating'])
    return df2

#第四小題
def directors_collaborate():
    director_dict={}
    for m in df.index:
        if df.iloc[m]['Director'] in director_dict.keys():
            for actor in df.iloc[m]['Actors']:
                if actor not in director_dict[df.iloc[m]['Director']]:
                    director_dict[df.iloc[m]['Director']].append(actor)
        else:
            director_dict[df.iloc[m]['Director']]=df.iloc[m]['Actors']
    for director in director_dict:
        director_dict[director]=len(director_dict[director])
    df3=pd.DataFrame(list(director_dict.items()))
    df3=df3.sort_values(1, ascending=False)
    return df3.head(3)

#第五小題
def actor_genre():
    file=pd.read_csv('IMDB-Movie-Data.csv')
    df=pd.DataFrame(file)
    df=df.fillna(0) 
    df['Genre']=df['Genre'].str.split('|')
    df['Actors']=df['Actors'].str.split('|')
    actor_genre_dict={}
    for m in df.index:
        for actor in df.iloc[m]['Actors']:
            if actor in actor_genre_dict.keys():
                for genre in  df.iloc[m]['Genre']:
                    if genre not in actor_genre_dict[actor]:
                        actor_genre_dict[actor].append(genre)
            else:
                actor_genre_dict[actor]=df.iloc[m]['Genre']
    for actor in actor_genre_dict:
        actor_genre_dict[actor]=len(actor_genre_dict[actor])
    df4=pd.DataFrame(list(actor_genre_dict.items()))
    df4=df4.sort_values(1, ascending=False)
    return df4.head(29)

#第六小題
def maximum_gap():
    actor_yaer_dict={}
    for m in df.index:
        for actor in df.iloc[m]['Actors']:
            if actor in actor_yaer_dict.keys():
                if df.iloc[m]['Year'] not in  actor_yaer_dict[actor]:
                    actor_yaer_dict[actor].append(df.iloc[m]['Year'])
                    actor_yaer_dict[actor].sort(reverse=True)
            else:
                actor_yaer_dict[actor]=[df.iloc[m]['Year']]
    for actor in actor_yaer_dict.keys():
        gap=actor_yaer_dict[actor][0]-actor_yaer_dict[actor][-1]
        actor_yaer_dict[actor]=gap
    df5=pd.DataFrame(list(actor_yaer_dict.items()))
    df5=df5.sort_values(1, ascending=False)
    return df5.head(147)

#第七小題
def collaboration(target = "Johnny Depp"):
    # Create undirected edges graph
    group = {}
    for m in df.index:
        names = df.iloc[m]['Actors']
        # Initialize/Check whether exists before
        for name in names:
            if name not in group.keys():
                group[name] = name
        for i in range(1,len(names)):
            union(names[0],names[i],group)
            
    ans_list = []
    for name in group:
        if find(name,group) == find(target,group) and name!=target:
            ans_list.append(name)
    df6=pd.DataFrame(ans_list,columns=['Johnny Depp'])
    return df6

    
def find(u,group):
    if u == group[u]:
        return u
    else:
        group[u] = find(group[u],group)
        return group[u]
    
def union(u,v,group):
    group[find(u,group)] = find(v,group)

print('第一題',highest_rating())
print('第二題',highest_average_revenue())
print('第三題',emma_average_rating())
print('第四題',directors_collaborate())
print('第五題',actor_genre())
print('第六題',maximum_gap())
print('第七題',collaboration(target = "Johnny Depp"))
```
