```python
#homework 0 第一題
def split(poly): #拆開多項式 以加減號分項
    symbol=[]
    last=0
    for i in range(len(poly)):
        if poly[i] == '+' or poly[i] == '-':
            symbol.append(poly[last:i])
            last=i
    symbol.append(poly[last:])
    if symbol[0][0] != '-':
        symbol[0]='+'+symbol[0]
    return symbol

def variable(poly): #判別變數XYZ
    var={}
    for i in range(len(split(poly))):
        for j in range(len(split(poly)[i])):
            if split(poly)[i][j].isalpha() == True:
                var[split(poly)[i][j]]=0
    return var

def coefficient(poly): #分出係數項
    coef1=[]
    for i in range(len(split(poly))):
        if split(poly)[i].find('*') != -1:
            coef=int(split(poly)[i][0:split(poly)[i].find('*')])
            coef1.append(coef)
        else:
            coef1.append(1)
    return coef1

def square(poly): #分出次方項 {x:0,y:0,z:0}
    square1=[]
    for i in range(len(split(poly))):
        square=variable(poly)
        if split(poly)[i].find('X')!= -1:
            if split(poly)[i][-1] == 'X' or split(poly)[i][split(poly)[i].find('X')+1] != '^':
                square['X']=1
            else:
                square['X']=int(split(poly)[i][split(poly)[i].find('X')+2])
        if split(poly)[i].find('Y')!= -1:
            if split(poly)[i][-1] == 'Y' or split(poly)[i][split(poly)[i].find('Y')+1] != '^':
                square['Y']=1
            else:
                square['Y']=int(split(poly)[i][split(poly)[i].find('Y')+2])
        if split(poly)[i].find('Z')!= -1:
            if split(poly)[i][-1] == 'Z' or split(poly)[i][split(poly)[i].find('Z')+1] != '^':
                square['Z']=1
            else:
                square['Z']=int(split(poly)[i][split(poly)[i].find('Z')+2])
        square1.append(square)
    return square1

#開始運算
def multiple(poly1,poly2):
    list_mul=[]
    for i in range(len(coefficient(poly1))):
        for j in range(len(coefficient(poly2))):
            mul=coefficient(poly1)[i]*coefficient(poly2)[j]
            list_mul.append(mul)
    return list_mul
def power(poly1,poly2):
    list_squ=[]
    for i in range(len(square(poly1))):
        for j in range(len(square(poly2))):
            dict_squ={}
            dict_squ['X']=square(poly1)[i]['X']+square(poly2)[j]['X']
            dict_squ['Y']=square(poly1)[i]['Y']+square(poly2)[j]['Y']
            if 'Z' not in square(poly1)[i]:
                dict_squ['Z']=square(poly2)[j]['Z']
            else:
                dict_squ['Z']=square(poly1)[i]['Z']+square(poly2)[j]['Z']
            list_squ.append(dict_squ)
    return list_squ

poly=input('Input the polynomial:')
symbol1=[] 
symbol2=[]
symbol3=[]
poly=poly.split(")(")
if len(poly)==2:
    poly1=poly[0].replace('(','') #poly1去掉括號
    poly2=poly[1].replace(')','') #poly2去掉括號
    

list1=list(zip(multiple(poly1,poly2),power(poly1,poly2)))
for i in range(len(list1)):
    if list1[i][1]['X']==0:
        if list1[i][1]['Y']==0:
            print(list1[i][0],'*','Z','^',list1[i][1]['Z'],sep='',end='+')
        else:
            print(list1[i][0],'*','Y','^',list1[i][1]['Y'],'Z','^',list1[i][1]['Z'],sep='',end='+')
    elif list1[i][1]['Y']==0:
        print(list1[i][0],'*','X','^',list1[i][1]['X'],'Z','^',list1[i][1]['Z'],sep='',end='+')
    elif list1[i][1]['Z']==0:
        print(list1[i][0],'*','X','^',list1[i][1]['X'],'Y','^',list1[i][1]['Y'],sep='',end='+')
    else:
        print(list1[i][0],'*','X','^',list1[i][1]['X'],'Y','^',list1[i][1]['Y'],'Z','^',list1[i][1]['Z'],sep='',end='+')
```
